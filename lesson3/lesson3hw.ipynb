# lesson3hw1

```python
import requests
from requests import Response
url = 'https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=5481753E-52AF-40DA-9A8A-9E192B245E13'

response: Response = requests.get(url)
print(response.text)

# 使用 with open() 將內容保存為 a1.csv
with open('a1.csv', 'w', encoding='utf-8') as file:
    file.write(response.text)

print("文件已保存為 a1.csv")
```

# lesson3hw2

```python
import requests
import json

# 定義URL
url = 'https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=44E49A04-8DE8-4E74-B872-BE322286C940'

# 下載文件
response = requests.get(url)

# 解析JSON內容
data = response.json()

with open('a2.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

    print("文件已成功下載並保存為a2.json")
```
