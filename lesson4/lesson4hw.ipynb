```python
from ttkthemes import ThemedTk
from tkinter import ttk


class Window(ThemedTk):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.title('甜點')

        style = ttk.Style(self)
        # 統一字體大小為 30
        style.configure('TButton', font=('Arial', 30))

        topFrame = ttk.Frame(self, borderwidth=3, relief='groove')
        topFrame.pack(padx=10, pady=(10, 0), expand=True, fill='x')

        # 創建上框架的按鈕
        btn1 = ttk.Button(topFrame, text="甜點1")
        btn1.pack(side='left', expand=True, fill='x', padx=10)

        btn2 = ttk.Button(topFrame, text="甜點2")
        btn2.pack(side='left', expand=True, fill='x', padx=10)

        btn3 = ttk.Button(topFrame, text="甜點3")
        btn3.pack(side='left', expand=True, fill='x', padx=10)

        # 創建下框架
        frame1 = ttk.Frame(self, borderwidth=3, relief='groove')
        frame1.pack(padx=10, pady=10, side='left', fill='y', expand=True)

        frame2 = ttk.Frame(self, borderwidth=3, relief='groove')
        frame2.pack(padx=10, pady=10, side='left', fill='y', expand=True)

        frame3 = ttk.Frame(self, borderwidth=3, relief='groove')
        frame3.pack(padx=10, pady=10, side='left', fill='y', expand=True)

        # 添加按鈕到下框架
        btn4 = ttk.Button(frame1, text='蛋糕')
        btn4.pack(fill='x', padx=10, pady=(5, 5))

        btn5 = ttk.Button(frame1, text='冰淇淋')
        btn5.pack(fill='x', padx=10, pady=(5, 5))

        btn6 = ttk.Button(frame1, text='布丁')
        btn6.pack(fill='x', padx=10, pady=(5, 5))

        btn7 = ttk.Button(frame2, text='馬卡龍')
        btn7.pack(fill='x', padx=10, pady=(5, 5))

        btn8 = ttk.Button(frame2, text='餅乾')
        btn8.pack(fill='x', padx=10, pady=(5, 5))

        btn9 = ttk.Button(frame2, text='泡芙')
        btn9.pack(fill='x', padx=10, pady=(5, 5))

        btn10 = ttk.Button(frame3, text='奶酪')
        btn10.pack(fill='x', padx=10, pady=(5, 5))

        btn11 = ttk.Button(frame3, text='巧克力')
        btn11.pack(fill='x', padx=10, pady=(5, 5))

        btn12 = ttk.Button(frame3, text='鬆餅')
        btn12.pack(fill='x', padx=10, pady=(5, 5))
        
        


def main():
    window = Window(theme="arc")
    window.mainloop()


if __name__ == '__main__':
    main()
```
